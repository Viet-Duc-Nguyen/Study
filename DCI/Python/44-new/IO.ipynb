```python
import io

with io.StringIO('Hello World') as s:
    s.seek(11)
    print(s.read())
    s.write('BLA')
    print(s.tell())
    print(s.read())
    print(s.getvalue())
    print(s.closed)
print(s.closed)
```

```python
import io

with io.StringIO() as f:
    f.write('first line\n')
    print('second line', file=f)
    print(f.getvalue())
```

```python
import io

import requests

res = requests.get(
    'https://images.unsplash.com/photo-1659523052948-4c82b042a7a6?ixlib=rb-4.0.3&ixid=MnwxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8&auto=format&fit=crop&w=2070&q=80')

with io.BytesIO() as buf:
    buf.write(res.content)
    buf.seek(0)
    post_res = requests.post('http://127.0.0.1:3000', files={"photo": buf})
    print(post_res)
```
